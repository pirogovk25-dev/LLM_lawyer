```python
%pip install python-docx langchain-text-splitters

import os
import json
import re
from docx import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

DATA_DIR = "./data"
PARSED_FILE = "parsed_documents.json"
CHUNKED_FILE = "chunked_documents.json"
```

```python
def clean_text(text):
    text = re.sub(r'\(в ред\. Федеральны[хго]+ закон[аов]+.*?\)', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def parse_docx(file_path):
    doc = Document(file_path)
    full_text = []
    
    for para in doc.paragraphs:
        cleaned_para = clean_text(para.text)
        if cleaned_para:
            full_text.append(cleaned_para)
            
    return "\n".join(full_text)

def run_parsing():
    documents = []
    for filename in os.listdir(DATA_DIR):
        if filename.endswith(".docx"):
            file_path = os.path.join(DATA_DIR, filename)
            print(f"Парсинг файла: {filename}...")
            text_content = parse_docx(file_path)
            documents.append({
                "source": filename,
                "content": text_content
            })
    
    with open(PARSED_FILE, 'w', encoding='utf-8') as f:
        json.dump(documents, f, ensure_ascii=False, indent=4)
    print(f"Готово! Обработано файлов: {len(documents)}. Результат в {PARSED_FILE}")

run_parsing()
```

```python
import random

def check_parsed_data():
    with open(PARSED_FILE, 'r', encoding='utf-8') as f:
        documents = json.load(f)
        
    print(f"Всего документов: {len(documents)}")
    random_doc = random.choice(documents)
    print(f"\nСлучайный фрагмент из {random_doc['source']}:")
    print(random_doc['content'][:500] + "...")

check_parsed_data()
```

```python
def run_chunking():
    with open(PARSED_FILE, 'r', encoding='utf-8') as f:
        documents = json.load(f)

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len,
        add_start_index=True,
    )

    all_chunks = []

    for doc in documents:
        source_name = doc["source"]
        text = doc["content"]
        chunks = text_splitter.split_text(text)
        
        for i, chunk_text in enumerate(chunks):
            all_chunks.append({
                "chunk_id": f"{source_name}_{i}",
                "source": source_name,
                "text": chunk_text
            })

    with open(CHUNKED_FILE, 'w', encoding='utf-8') as f:
        json.dump(all_chunks, f, ensure_ascii=False, indent=4)

    print(f"Из {len(documents)} документов создано {len(all_chunks)} смысловых чанков.")
    print(f"Результат сохранен в: {CHUNKED_FILE}")

run_chunking()
```
